### 6. Membuat bentuk Geometri 2D dari OpenCV.Diawali dengan pembuatan black image dengan tipe data int16.

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
black_img = np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(black_img)
```

```python
#Menambahkan bentuk persegi panjang sesuai dengan koordinat pt1 dan pt2

cv.rectangle(black_img,pt1=(384,0),pt2=(510,150),color=(0,255,0),thickness=10)
plt.imshow(black_img)
```

```python
#menambahkan bentuk persegi sesuai koordinat pt1 dan pt2 yang tertulis pada kode program.
cv.rectangle(black_img, pt1=(200,200), pt2=(300,300), color=(0,0,255), thickness=15)
plt.imshow(black_img)
```

```python
#menambahkan bentuk lingkaran sesuai radius yang tertulis pada kode program
cv.circle(black_img, center=(100,100), radius=50, color=(255,0,0), thickness=8)
plt.imshow(black_img)
```

```python
# Menambahkan garis sesuai koordinat pt1 dan pt2
cv2.circle(black_img,center=(400,400),radius=50,color=(255,0,0),thickness=-1)
cv2.line(black_img,pt1=(0,0),pt2=(512,512),color=(255,255,255),thickness=5)
plt.imshow(black_img)
```

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(
    black_img,
    text='Hello',
    org=(10,500),
    fontFace=font,
    fontScale=4,
    color=(255,255,0),
    thickness=2,
    lineType=cv.LINE_AA
)
plt.imshow(black_img)
```

```python
black_img2 = np.zeros(shape=(512,512,3), dtype=np.int32)
plt.imshow(black_img2)
```

```python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]], dtype=np.int32)
pts = vertices.reshape((-1,1,2))  # nilai 2 menunjukkan bahwa tiap titik dibuat berisi pasangan (x,y)
pts
```

```python
cv.polylines(black_img2,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(black_img2)
```
